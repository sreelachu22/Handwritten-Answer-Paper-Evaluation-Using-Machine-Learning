 Keyword extraction


```python
from rake_nltk import Rake
def keywordextract(answer):
  r = Rake()

# Extraction from the text.
  r.extract_keywords_from_text(answer)
  L=[]
# To get keywords
  L=r.get_ranked_phrases()
  return L
```

 Synonym generation


```python
from nltk.corpus import wordnet
def synonymgen(answer) :
  synonyms = []
  L=keywordextract(answer)
  print("Keywords",L)
  noofkey=len(L)
  return L,noofkey
  for i in L:
    synonyms.append(i)
    for syn in wordnet.synsets(i): #similar words are grouped into synset
      for lemma in syn.lemmas(): #words in a Synset are known as Lemmas
        synonyms.append(lemma.name())
  #for a in synonyms:
    #print(a)
    synonyms = list( dict.fromkeys(synonyms) ) #to remove dup
```

Keywords stored in file keyword-stud.txt 

```python
keywords = open("/content/drive/MyDrive/keyword-rake/keywords.txt","r")
file1 = open("/content/drive/MyDrive/keyword-rake/stud-answer.txt","r")
file2 = open("/content/drive/MyDrive/keyword-rake/keyword-stud.txt","w")
for k in range(1,no_of_qs+1):
  answer=file1.readline()
  #print(answer) #student's answer
  
  synonyms,noofkey=synonymgen(answer) 
  for j in synonyms:
      file2.write(j+"*")
  file2.write("\n")
file1.close()
file2.close()
```
